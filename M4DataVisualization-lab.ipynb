```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
df['WorkWeekHrs'].median()
```

```python
df['CodeRevHrs'].median()
```

```python
# Aplica el filtro primero
filtro = (df['Age'] >= 30) & (df['Age'] <= 35)
df_filtrado = df[filtro]

# Agrupa por una columna, por ejemplo, 'Age' y calcula la mediana de 'WorkWeekHrs'
agrupado_WorkWeekHrs = df_filtrado.groupby('Age')['WorkWeekHrs'].median()

agrupado_CodeRevHrs = df_filtrado.groupby('Age')['CodeRevHrs'].median()
agrupado_WorkWeekHrs
```

```python
agrupado_CodeRevHrs
```

```python
# your code goes here


weight_counts = {
    "WorkWeekHrs": agrupado_WorkWeekHrs.values,
    "CodeRevHrs": agrupado_CodeRevHrs.values,
}

width = 0.5
fig, ax = plt.subplots()
bottom = np.zeros(len(agrupado_WorkWeekHrs))

for label, weight_count in weight_counts.items():
    p = ax.bar(agrupado_WorkWeekHrs.index, weight_count, width, label=label, bottom=bottom)
    bottom += weight_count

ax.set_title("Distribución de horas según la edad")
ax.legend(loc="upper right")
plt.xlabel("Age")
plt.ylabel("Hours")
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

filtro = (df['Age'] >= 45) & (df['Age'] <= 60)
df_filtrado2 = df[filtro]
converted40 = df_filtrado2.groupby('Age')['ConvertedComp'].median()

sns.relplot(x=converted40.index, y=converted40.values, kind='line')
plt.xlabel("Age")
plt.ylabel("Median ConvertedComp")
plt.title("Median Converted Compensation for Ages 45 to 60")
plt.grid()
plt.show()
```
